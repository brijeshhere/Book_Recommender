# Collaborative Filtering based Recommender System

```python
active_user=books_rate_and_name.groupby('User-ID').count()['Book-Rating']
```

```python
active_user=active_user[active_user>50]
```

```python
active_user
```

```python
filtered_df=books_rate_and_name[books_rate_and_name['User-ID'].isin(active_user)]
filtered_df.head(3)
```

```python
famous_books=books_rate_and_name.groupby('Book-Title').count()['Book-Rating']
famous_books
```

```python
famous_books=famous_books[famous_books>50]
```

```python
famous_books
```

```python
filtered_df
```

```python
filtered_df['Book-Title'].isin(famous_books).sum()
```

```python
final_rating=filtered_df[filtered_df['Book-Title'].isin(famous_books.index)]
```

```python
final_rating.head()
```

```python
pt=final_rating.pivot_table(index=['Book-Title'],columns=['User-ID'],values=['Book-Rating'],fill_value=0)
```

```python
pt
```

```python
pt=pt.droplevel(0,axis=1)
```

```python
pt.sample(20)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
pt.shape
```

```python
similar_score=cosine_similarity(pt)
```

```python
np.where(pt.index=='1984')
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar=sorted(list(enumerate(similar_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    l=[]
    for i in similar:
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        temp_df=temp_df.drop_duplicates(subset=['Book-Title'])
        temp_df=temp_df[['Book-Title','Book-Author','Image-URL-M']]
        l.append(temp_df.to_numpy()[0])
    return l
    
```

```python
recommend('1984')
```
